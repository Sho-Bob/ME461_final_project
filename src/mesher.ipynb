```python
import numpy as np
import matplotlib.pyplot as plt

# Generate mesh grid in 3D space (x, y, z)

Nx = 256; Ny = 128; Nz = 128

# y and z are uniform grids
y = np.linspace(0,2.0*np.pi,Ny)
z = np.linspace(0,2.0*np.pi,Nz)

# x is a non-uniform grid, 
x = np.linspace(0,2.0*np.pi,Nx)


# Print data of mesh
print("+++++++++++++++++++++++++++")
print("Mesh data:")
print("x points: ", len(x))
print("y points: ", len(y))
print("z points: ", len(z))
print("x domain size: ", x[0], "To", x[-1])
print("y domain size: ", y[0], "To", y[-1])
print("z domain size: ", z[0], "To", z[-1])
print("+++++++++++++++++++++++++++")

# Save x, y, z to separate .dat files
np.savetxt('x.dat', x, fmt='%.18e')
np.savetxt('y.dat', y, fmt='%.18e')
np.savetxt('z.dat', z, fmt='%.18e')

print(f"Saved x ({len(x)} points) to x.dat")
print(f"Saved y ({len(y)} points) to y.dat")
print(f"Saved z ({len(z)} points) to z.dat")

```
